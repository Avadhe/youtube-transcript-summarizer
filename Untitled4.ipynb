```python
from youtube_transcript_api import YouTubeTranscriptApi

# Video ID from the YouTube URL
video_id = "82P5N2m41jE&list=PLKnIA16_Rmvbr7zKYQuBfsVkjoLcJgxHH&index=13"

try:
    # Fetch the transcript in Hindi (language code 'hi')
    transcript = YouTubeTranscriptApi.get_transcript(video_id, languages=['hi'])

    # Print out the transcript
    for entry in transcript:
        print(f"{entry['start']} - {entry['duration']}: {entry['text']}")

except Exception as e:
    print(f"Error fetching transcript: {str(e)}")
```
